```python
def write_wav_file(path:str, data:np.ndarray, sample_rate:int)->None:
    wavfile.write(path, sample_rate, data)
```

```python
# extract 10-sec utterances with all fillers
filenames=('31-V-C-A.wav','31-V-NC-A.wav','32-V-C-B.wav', '33-V-C-A.wav', '34-V-C-B.wav', '40-V-NC-B.wav', '41-V-NC-A.wav', '44-V-NC-B.wav')
timesteps_arr=((129, 139), (170, 180), (307, 317), (23, 33), (30, 40), (76, 86), (200, 210), (267, 277))
filler_types=('2_m','2_m','2_mhm', '2_ja', '2_mhm', '2_m', '2_m', '2_m')
timesteps_to_cut_arr=(())


for filename, timesteps, filler_type in zip(filenames, timesteps_arr, filler_types):
    print(filename)
    timesteps=timesteps
    additional_interval=0
    path_to_file=os.path.join(r'D:\Databases\ALICO\ALICO\recordings', filename)
    filler_type=filler_type
    wav_file, sample_rate, _=extract_utterance_according_timesteps(path_to_file=path_to_file, timesteps=timesteps, additional_interval=additional_interval)

    path_to_save=path_to_file.split('\\')[-1].split('.')[0]+'_%.2f_%.2f_%s.wav'%(timesteps[0], timesteps[1], filler_type)
    write_wav_file(path=path_to_save, data=wav_file, sample_rate=sample_rate)
    #IPython.display.Audio(path_to_save)
```

```python
# extract 10-sec utterances with one cut filler
filenames=('31-V-C-A.wav','31-V-NC-A.wav','32-V-C-B.wav', '33-V-C-A.wav', '34-V-C-B.wav', '40-V-NC-B.wav', '41-V-NC-A.wav', '44-V-NC-B.wav')
timesteps_arr=((129, 139), (170, 180), (307, 317), (23, 33), (30, 40), (76, 86), (200, 210), (267, 277))
filler_types=('2_m','2_m','2_mhm', '2_ja', '2_mhm', '2_m', '2_m', '2_m')
timesteps_to_cut_arr=((130.304,130.611),(173.104,173.354),(310.972,311.469), (27.482,27.698),(31.388,31.731),(84.759,85.062),(202.312,202.482),(274.228,274.478))


for filename, timesteps, filler_type, timesteps_to_cut in zip(filenames, timesteps_arr, filler_types, timesteps_to_cut_arr):
    print(filename)
    timesteps=timesteps
    additional_interval=0
    path_to_file=os.path.join(r'D:\Databases\ALICO\ALICO\recordings', filename)
    filler_type=filler_type+'_one_cut'
    wav_file, sample_rate, _=extract_utterance_according_timesteps(path_to_file=path_to_file, timesteps=timesteps, additional_interval=additional_interval)
    wav_file=wav_file.copy()
    start_cut=int((timesteps_to_cut[0]-timesteps[0])*sample_rate)
    end_cut=int((timesteps_to_cut[1]-timesteps[1])*sample_rate)
    wav_file[start_cut:end_cut,1]=0
    wav_file[start_cut:end_cut,0]=0
    path_to_save=path_to_file.split('\\')[-1].split('.')[0]+'_%.2f_%.2f_%s.wav'%(timesteps[0], timesteps[1], filler_type)
    write_wav_file(path=path_to_save, data=wav_file, sample_rate=sample_rate)
    #IPython.display.Audio(path_to_save)
```
